## Cleaning the data

```python
# rename the first column as "carname"

cars_df.rename({'Unnamed: 0':'carname'}, axis = 1, inplace = True)
cars_df.head()
```

```python
# rename the mpg as "MilesPerGallon", cyl as "NumOfCylinders"

cars_df.rename({'mpg':'MilesPerGallon','cyl':'NumOfCylinders'}, axis = 1, inplace = True)
cars_df.head()
```

```python
# replace Datsum 710 with Datsun Go 710
cars_df['carname'].replace({'Datsun 710':'Datsun Go 710'},inplace = True)
cars_df.head()
```

```python
cars_df['NumOfCylinders'].replace({6:7})
```

```python
cars_df.head()
```

```python
cars_df_cleaned = cars_df.drop(['carb','gear'], axis = 1) # inorder to make this drop permanent in the data frame, mention inplace = True
cars_df_cleaned
```

```python
# locating a specific value in the data frame. functions loc()-> locate based on rows nad column reference
#  and iloc()--> index loction

cars_df['hp'].iloc[17]
```

```python
# sort the data based on the miles per galon in ascending order
cars_df.sort_values(by = 'MilesPerGallon', ascending = True, ignore_index = True)#, inplace = True)
# sort in descending order
#cars_df.sort_values(by = 'MilesPerGallon', ascending = False, ignore_index = True)
```

```python
# sorting the data using multiple columns
cars_df.sort_values(by = ['MilesPerGallon','gear'], ascending = [False, True], ignore_index = True)#, inplace = True)
```

```python
# getting input from the user

def avg_marks():
  mark1 = input("Enter your mark1: ")
  mark2 = input("Enter your mark2: ")
  mark3 = input("Enter your mark3: ")
  return round((int(mark1)+int(mark2)+int(mark3))/3,2)

avg_marks()
```
